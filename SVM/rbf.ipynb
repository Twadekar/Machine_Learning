```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mushroom.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
colname = []
for x in df.columns:
    if df[x].dtypes=="object":
        colname.append(x)
colname
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for x in colname:
    df[x] = le.fit_transform(df[x])

    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print("features", x)
    print("mapping", le_name_mapping)
```

```python
plt.figure(figsize=(30, 6))
df.boxplot()
plt.show()
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

df_new = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR ))).any(axis=1)]
```

```python
print("original shape: ", df.shape)
print("shape after removing outliers: ", df_new.shape)
```

```python
x = df[["cap_shape", "cap_surface", "cap_color", "bruises", "odor", "gill_attachment", "gill_spacing", "gill_size", "gill_color", "stalk_shape",
         "veil_type", "veil_color", "ring_number", "ring_type", "spore_print_color", "population", "habitat", "stalk_height", "cap_diameter"]]
y = df[["class"]]
```

```python
print(x.shape)
print(y.shape)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x, y = smote.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
modelr = RandomForestClassifier()
modelr.fit(x_train, y_train)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf')
model.fit(x_train, y_train.values.ravel())
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score

y_pred = model.predict(x_test)

acc= accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

cr = accuracy_score(y_test, y_pred)
print("classification_report", cr)

cm = confusion_matrix(y_test, y_pred)
print("confusion metrix:", cm)

fs = f1_score(y_test, y_pred)
print("confusion metrix:", fs)
```

```python
#visualization of testing data

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
x_test_2d = pca.fit_transform(x_test)

```
