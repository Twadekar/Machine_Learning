```python
print(df["sex"].unique())
print(df["exang"].unique())
```

```python
df["exang"] = df["exang"].replace({
    'False' : 'FALSE',
    'TURE' : 'TRUE',
    'True' : 'TRUE'
})
```

```python
df.dtypes
```

```python
df['exang'] = df['exang'].astype(str).str.lower().str.strip()
```

```python
#Label Encoding
df['sex'] = df['sex'].map({'Male':1, 'Female':0})
df['exang'] = df['exang'].map({'false':1, 'true':0})
```

```python
df['num'] = df['num'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df["sex"].unique()
```

```python
df["exang"].unique()
```

```python
#One hot coding
df = pd.get_dummies(df, columns=["cp", "fbs", "restecg","slope","thal"], drop_first=True)
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('-', '_')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.astype({col: 'float' for col in df.select_dtypes(bool).columns})
df = df.astype({col: 'float' for col in df.select_dtypes(int).columns})
```

```python
df.dtypes
```

```python
x = df[['age', 'sex', 'trestbps', 'chol', 'thalch', 'exang', 'oldpeak',
       'cp_atypical_angina', 'cp_non_anginal', 'cp_typical_angina', 'fbs_True',
       'restecg_normal', 'restecg_st_t_abnormality', 'slope_flat',
       'slope_upsloping', 'thal_normal', 'thal_reversable_defect']]
y= df["num"]
```

```python
df.dtypes
```

```python
import joblib
feature_names= x.columns
joblib.dump(feature_names, "features.pkl")
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x , y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(class_weight='balanced', random_state=42)
clf.fit(x_train, y_train)
```

```python
#Import model DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion="entropy", max_depth=7, random_state=42, min_samples_split=15, min_samples_leaf=10)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Accuracy:", accuracy_score(y_test, y_pred))
print("classification_report:", classification_report(y_test, y_pred))
print("confusion_matrix", confusion_matrix(y_test, y_pred))
```

```python
unique_classes = clf.classes_
class_names_n = [str(c) for c in unique_classes]
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(30, 17))
plot_tree(model, feature_names = x.columns, class_names=class_names_n, filled=True, rounded=True, fontsize=7)
plt.show()
```

```python
import pickle
with open("model.pkl", "wb") as p:
    pickle.dump(model, p)
```
